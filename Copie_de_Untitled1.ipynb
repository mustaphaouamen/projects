ce code génère un graphique à barres pour visualiser les ventes par région,
identifiant ainsi la ville avec le plus grand nombre de ventes dans le jeu de données.
 Le graphique est personnalisé avec des titres, des étiquettes d'axe et un style spécifiques

```python

#Quelle ville a enregistré le plus grand nombre de ventes?
# groupe du plus grand nombre de ventes en ville
highest_number_of_df = df.groupby('Region').sum().astype('int')
# let's plot it
plt.figure(figsize=(15, 8)) # figuring the size

# makes bar plot
sns.barplot( # barplot
    x=highest_number_of_df.index, # x-axis
    y=highest_number_of_df["Sales"], # y-axis
    data=highest_number_of_df, # data
    palette="deep" # palette
)
plt.title( # title
    "City with the Highest number of Sales",
    fontname="monospace", # font-name
    weight="bold", # weight
    fontsize=35, # font-size
    pad=30 # padding
)
plt.xlabel( # x-label
    "Cities",
    weight="bold", # weight
    color="purple", # color
    fontsize=25, # font-size
    loc="center" # location
)
plt.xticks( # x-ticks
    weight="bold", # weight
    fontsize=15, # font-size
    rotation=10
)
plt.ylabel( # y-label
    "Sales in USD ($)",
    weight="bold", # weight
    color="green", # color
    fontsize=20 # font-size
)
plt.yticks( # y-ticks
    weight="bold", # weight
    fontsize=15 # font-size
);
```

```python
#Quels produits sont le plus souvent vendus ensemble?
from itertools import combinations
from collections import Counter
#le supprimer en utilisant la fonction duplicated()
data = df[df['Order_ID'].duplicated(keep=False)]
#créer une nouvelle colonne
data['Grouped'] = df.groupby('Order_ID')['Product_Category'].transform(lambda x: ','.join(x))
# créer une nouvelle variable
data = data[['Order_ID', 'Grouped']].drop_duplicates()
# créer une nouvelle variable pour le compteur
count = Counter()
# faire une boucle
for row in data['Grouped']:
  row_list = row.split(',')
  count.update(Counter(combinations(row_list, 2)))
#et en faire une autre boucle
for key, value in count.most_common(10):
    print(key, value)
```

```python
#Quels produits sont le plus souvent vendus ensemble?
from itertools import combinations
from collections import Counter
# Supprimer les lignes dupliquées basées sur 'Order_ID'
data = df[df['Order_ID'].duplicated(keep=False)]
# Créer une nouvelle colonne 'Grouped' avec les catégories de produits regroupées par commande
data['Grouped'] = df.groupby('Order_ID')['Product_Category'].transform(lambda x: ','.join(x))
# Créer une nouvelle variable avec les paires de produits différents
count = Counter()
# Boucle sur les groupes de produits dans chaque commande
for row in data['Grouped']:
    row_list = row.split(',')
    # Utiliser combinations pour obtenir toutes les paires possibles
    pairs = combinations(row_list, 2)
    # Filtrer les paires pour ne compter que les paires de produits différents
    different_product_pairs = [(product1, product2) for product1, product2 in pairs if product1 != product2]
    # Mettre à jour le compteur
    count.update(different_product_pairs)
# Afficher les résultats
for key, value in count.most_common(10):
    print(key, value)
```
